<h2> <span style="color:#32AD94"> Relational Scatter Plot </span> </h2>


<h3 style = "color:#AD4799">  Tips Dataset  </h3>

```python
import seaborn as sns 
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
tips.head()
```

<h3 style="color:#1F84D7"> Parameters = row, row_order, col, col_order, kind, hue, style</h3>
<h4 style="color:#1F84D7"> This plot shows the relationship between total_bill and tips grouped by smoker and time of eating </h4>

```python
sns.set_style("whitegrid")
sns.set_context("talk")
sns.relplot(x="tip",
            y="total_bill",
            data=tips,kind="scatter",
            row="smoker",col="time",
            col_order=['Lunch','Dinner'],
            row_order=['Yes','No'],
            size="size",
            hue="size",
           style="size",
           palette="icefire")
plt.show()
```

<h2> <span style="color:#32AD94"> Relational Line Plot </span> </h2>


<h3 style = "color:#AD4799">  MPG Dataset  </h3>

```python
mpg =sns.load_dataset("mpg")
mpg.head()
```

<h3 style="color:#1F84D7"> Parameters = style, hue, ci, dashes, markers</h3>
<h4 style="color:#1F84D7"> This plot shows the relationship between Cylinders and MPG grouped by origin country </h4>

```python
sns.set_context("talk")
sns.set_style("whitegrid")
plot=sns.relplot(data=mpg,
            kind="line",
            x="cylinders",
            y="mpg",
            style="origin",
            hue="origin",
            ci=None,
            dashes=False,
            markers=True,
           palette="magma")
plot.fig.suptitle("MPG of Cylinders by Origin Country",y=1.1)
plt.show()
```

<h2> <span style="color:#32AD94"> Categorical Count Plot </span> </h2>


<h3 style = "color:#AD4799">  Titanic Dataset  </h3>

```python
titanic =sns.load_dataset("titanic")

titanic.head()
```

<h3 style="color:#1F84D7"> Parameters = hue, order</h3>


```python
sns.set_context("talk")
sns.set_style("whitegrid")
plot=sns.catplot(data=titanic,
            kind="count",
            x="survived",
            hue="who",
            hue_order=['child','woman','man'],
            order=[1,0],
            palette="mako")
plot.fig.suptitle("Survival Count by Child, Man or Woman",y=1.1)
plt.show()
```

<h2> <span style="color:#32AD94"> Categorical Bar Plot </span> </h2>

```python
sns.set_context("talk")
plot=sns.catplot(kind="bar",
            data=titanic,
            x="embark_town",
            y="fare",
           palette="rocket")
plot.fig.suptitle("Embarked Town vs Fare",y=1.1)
plt.xticks(rotation=45) 
plt.show()
```

<h2> <span style="color:#32AD94"> Categorical Box Plot </span> </h2>

<h4 style="color:#1F84D7"> Box plot shows total_bill by day  </h4>
<p style="color:#1F84D7"> Median = 50th percentile, Interquartile range is between 25th and 75th percentile, whisker is between 5th and 95th percentile</p>   

```python
sns.set_context("talk")
plot=sns.catplot(kind="box",data=tips,x="day",y="total_bill",whis=[5,95],sym="",palette="husl")
plot.fig.suptitle("Total bill by Day of Week",y=1.1)
plt.xticks(rotation=45) 
plt.show()
```

<h2> <span style="color:#32AD94"> Categorical Point Plot </span> </h2>

```python
import numpy as np
sns.set_context("talk")
plot=sns.catplot(kind="point",
            data=titanic,
           x="who",
           y="fare",
           join=False,
           capsize=0.2,
           estimator=np.median,
           palette="Set2")
plot.fig.suptitle("Fare by Child, Man or Woman",y=1.1)
plt.xticks(rotation=45)         
plt.show()
```
